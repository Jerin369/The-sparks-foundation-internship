```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
df = pd.read_csv('iris.csv',index_col=0)
df.head()
```

```python
data = df.drop(['Species'],axis=1)
data
```

```python
from sklearn.cluster import KMeans
wcss=[]
k_rang = 10
for i in range(1,k_rang):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(data)
    wcss.append(kmeans.inertia_)
    
clusters=range(1,10)
plt.plot(clusters,wcss)
plt.title('Visualization of raw data');
```

```python
plt.scatter(data.iloc[:, 0], data.iloc[:, 1])
```

```python
centroids = kmeans.cluster_centers_
# Plot the clustered data
fig, ax = plt.subplots(figsize=(6, 6))

plt.scatter(data['SepalLengthCm'], data['SepalWidthCm'],c='green', label='cluster 1')

plt.scatter(data['PetalLengthCm'], data['PetalWidthCm'],
            c='blue', label='cluster 2')

plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', s=300,
            c='r', label='centroid')
```

```python
 kmeans2 = KMeans(n_clusters = 2)
 kmeans2.fit(data)
 pred = data.copy()
 pred['predicted'] = kmeans2.fit_predict(data)
 pred
 plt.figure(figsize=(6, 6))

plt.scatter(pred['SepalLengthCm'], pred['SepalWidthCm'],c=pred['predicted'], label='cluster 1')
plt.title('SepalLength vs SepalWidth')
pred
```

```python
plt.scatter(pred['PetalLengthCm'], pred['PetalWidthCm'],c=pred['predicted'], label='cluster 1')
plt.title('PetalLength vs PetalWidth')
```

```python
 kmeans3 = KMeans(n_clusters = 3)
 kmeans3.fit(data)
 pred2 = data.copy()
 pred2['predicted'] = kmeans3.fit_predict(data)
 pred2
 plt.figure(figsize=(6, 6))

plt.scatter(pred2['SepalLengthCm'], pred2['SepalWidthCm'],c=pred2['predicted'], label='cluster 1')
plt.title('SepalLength vs SepalWidth')
```

```python
plt.scatter(pred2['PetalLengthCm'], pred2['PetalWidthCm'],c=pred2['predicted'], label='cluster 1')
plt.title('PetalLength vs PetalWidth')
```
